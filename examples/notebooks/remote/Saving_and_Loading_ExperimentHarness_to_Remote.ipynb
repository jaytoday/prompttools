```python
from prompttools.experiment import OpenAIChatExperiment
from prompttools.harness import SystemPromptExperimentationHarness, ChatPromptTemplateExperimentationHarness, ModelComparisonHarness

import os

os.environ["OPENAI_API_KEY"] = ""
os.environ["HEGELAI_API_KEY"] = ""
```

## Prompt Template Harness

```python
message_templates = [
    [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Who was the {{input}} president?"},
    ],
    [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Who was the {{input}} vice president?"},
    ]
]

user_inputs = [{"input": "first"}, {"input": "second"}]


harness = ChatPromptTemplateExperimentationHarness(OpenAIChatExperiment,
                                                   "gpt-3.5-turbo",
                                                   message_templates,
                                                   user_inputs,
                                                   model_arguments=None)
```

```python
harness.prepare()
harness.run()
harness.visualize()
```

### Examples of `aggregation` API

```python
agg_df = harness.aggregate(groupby_column='user_inputs', aggregate_columns='latency', method='mean')
agg_df
```

```python
agg_df = harness.aggregate(groupby_column='templates', aggregate_columns='latency', method='mean')
agg_df
```

```python
agg_df = harness.aggregate(groupby_column='templates', aggregate_columns=['latency', 'latency'], method='mean')
agg_df
```

```python
harness.save_experiment("HistoryExam")
```

### Load from Remote

```python
harness_from_load = ChatPromptTemplateExperimentationHarness.load_experiment(harness._experiment_id)
```

```python
harness_from_load.visualize()
```

```python
harness_from_load.run(clear_previous_results=True)  # You can set to `False` to persist previous results
harness_from_load.visualize()
```

## Model Comparison

```python
model_names = ['gpt-3.5-turbo', 'gpt-3.5-turbo']
system_prompts = ["You are a helpful assistant.", "You are a 5th grade history teacher."]
user_messages = ["Who was the first president?"]

harness = ModelComparisonHarness(model_names, system_prompts, user_messages)
```

```python
harness.run()
harness.visualize()
```

```python
def _simple_eval(row):
    return "EVAL'ED " + row['response'][:10]


harness.evaluate("simple_eval_metric", _simple_eval)
harness.visualize()
```

```python
harness.save_experiment("HistoryModelComparison")
```
